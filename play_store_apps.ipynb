```python
from bs4 import BeautifulSoup
from selenium import webdriver 
import pandas as pd
import re
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://play.google.com/store/apps")
response = driver.execute_script("return document.documentElement.outerHTML")
```

```python
soup = BeautifulSoup(response, 'html.parser')
apps = soup.find_all("div",class_="WsMG1c nnK0zc")
name_app = []
links = []

div_link_app = soup.find_all("div",class_="vU6FJ p63iDd")

for app in apps:
    name_app.append(app.text)

for div in div_link_app:
    a = div.find('a', href = True)
    if a is not None:
        links.append("https://play.google.com"+a['href'])
```

```python
links
```

```python
apps_data = pd.DataFrame()
```

```python
apps_data['app'] = name_app
```

```python
apps_data['link'] = links
```

```python
apps_data
```

```python
rated = []
installs = []
genders= []

for a in links:
    driver.get(a)
    response = driver.execute_script("return document.documentElement.outerHTML")
    soup = BeautifulSoup(response, 'html.parser')
    rate = soup.find_all("div",class_="BHMmbe")
    install = soup.find('span', {'class' : 'AYi5wd TBRnV'}).get_text()
    gender = soup.find_all('a', {'class' : 'hrTbp R8zArc'})[-1].get_text()
    
    
    
    for rt in rate:
        rt_nuevo = rt.text.replace(',', '.')
        rated.append(float(rt_nuevo))
    
    installs.append(install)
    genders.append(gender)
```

```python
apps_data['rated'] = rated
apps_data['installs'] = installs
apps_data['gender'] = genders
```
