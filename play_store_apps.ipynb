```python
from bs4 import BeautifulSoup
from selenium import webdriver 
import pandas as pd
import re
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://play.google.com/store/apps")
response = driver.execute_script("return document.documentElement.outerHTML")
```

```python
soup = BeautifulSoup(response, 'html.parser')
apps = soup.find_all("div",class_="WsMG1c nnK0zc")
name_app = []
links = []

div_link_app = soup.find_all("div",class_="vU6FJ p63iDd")

for app in apps:
    name_app.append(app.text)

for div in div_link_app:
    a = div.find('a', href = True)
    if a is not None:
        links.append("https://play.google.com"+a['href'])
```

```python
links
```

```python
apps_data = pd.DataFrame()
```

```python
apps_data['app'] = name_app
```

```python
apps_data['link'] = links
```

```python
apps_data
```

```python
rated = []
valorations = []
genders= []

for a in links:
    driver.get(a)
    response = driver.execute_script("return document.documentElement.outerHTML")
    soup = BeautifulSoup(response, 'html.parser')
    rate = soup.find_all("div",class_="BHMmbe")
    valoration = soup.find('span', {'class' : 'AYi5wd TBRnV'}).get_text()
    gender = soup.find_all('a', {'class' : 'hrTbp R8zArc'})[-1].get_text()
    
    
    
    for rt in rate:
        rt_nuevo = rt.text.replace(',', '.')
        rated.append(float(rt_nuevo))
    
    valorations.append(valoration)
    genders.append(gender)
```

```python
apps_data['rated'] = rated
apps_data['valorations'] = valorations
apps_data['gender'] = genders
```

```python
'''
0: Last update date
1: Size MB-GB
2: Installs
3: Current version
4: Requires Android
5: Supported Version
6: Classification
7: App elements
8: Integrated app elements price
9: Permissions /x
10: Report x/
11: Company

'''
driver1 = webdriver.Chrome(ChromeDriverManager().install())


'''

driver1.get("https://play.google.com/store/apps/details?id=com.facebook.katana")
others = driver1.find_elements_by_class_name("htlgb")
for x in range (len(others)):
     if x % 2 == 0:
            list_others.append(others[x].text)
print(list_others)
list_others[10]
'''
list_others = []
lista = []

for a in links:
    driver1.get(a)
    others = driver1.find_elements_by_class_name("htlgb")
    list_aux = []
    
    for x in range (len(others)):
        if x % 2 == 0:
            list_aux.append(others[x].text)
    lista.append(list_aux)
print(lista)
```

```python
last_update = []
size = []
installs = []
current_version_app = []
android_version_req = []
classification = []

for a in lista:
    aux_last_update = a[0]
    aux_size = a[1]
    aux_installs = a[2]
    aux_current_version_app = a[3]
    aux_android_version_req = a[4]
    aux_classification = a[5]
    aux_company = a[8]
    
    last_update.append(a[0].strip("\n"))

    # Transformando variable size 
    if a[1] == 'Varía con el dispositivo.':
        aux_size = '0'
    if 'M' in a[1]:
        aux_size = a[1].strip('M')

    size.append(float(aux_size))
    
    # Transformando variable installs
    aux_installs = aux_installs.strip('+').replace(",", "")
    installs.append(int(aux_installs))

    # Transformando variable current_version_app
    aux_current_version_app = aux_current_version_app.strip('\n')
    current_version_app.append(aux_current_version_app)
    
    # Transformando variable android_version_req
    aux_android_version_req = aux_android_version_req.strip('\n')
    aux_android_version_req = aux_android_version_req.strip(' y versiones posteriores')
    android_version_req.append(aux_android_version_req)
    
    # Transformando variable classification
    
    aux_classification_lista = aux_classification.split('\n')
    classification.append(aux_classification_lista[0])
    
```

```python
apps_data['last_update'] = last_update
apps_data['size'] = size
apps_data['installs'] = installs
apps_data['current_version_app'] = current_version_app
apps_data['android_version_req'] = android_version_req
apps_data['classification'] = classification
```
